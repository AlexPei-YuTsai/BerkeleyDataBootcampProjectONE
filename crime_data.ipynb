Crime Data Analysis (from Federal Bureau of Investigation Crime Data Explorer)

Create DataFrame for Total Crimes from imported CSV File.

```python
#Dependencies

import numpy as np
import matplotlib.pyplot as plt

import pandas as pd 
from pathlib import Path


#Upload CSV

crime_data_load = Path("estimated_crimes_1979_2020.csv")

crime_data = pd.read_csv(crime_data_load)

crime_data
```

Clean Data to only show the relevant data we want in the DataFrame (Year, State, Total Crimes)

```python
crime_summary = crime_data[["year","state_name","total_crimes"]]
crime_summary
```

Extract Data to Find the number of Total Crimes per State from 1979 to 2020

```python
per_state_crimes = crime_summary.groupby(["state_name"])["total_crimes"].sum()
per_state_crimes
```

Extract Data to find the number of Total Crimes per Year from 1979 to 2020

```python
per_year_crimes_df = crime_summary.groupby(["year"])["total_crimes"].sum()
per_year_crimes_df
```
