# State Wage

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

# File to Load (Remember to Change These)
wage_estimates_to_load = Path("output/May 2022 State Occupational Employment and Wage Estimates.csv")


# Read School and Student Data File and store into Pandas DataFrames
wage_estimates = pd.read_csv(wage_estimates_to_load)

wage_estimates.head()
```

---

### Step 1:

```python
# Remove unwanted columms
wage_estimates = wage_estimates.drop(['AREA_TYPE', 'NAICS', 'O_GROUP', 'TOT_EMP', 'JOBS_1000','PCT_TOTAL', 'MEAN_PRSE', 'H_PCT10', 'H_PCT25', 'PRIM_STATE', 'EMP_PRSE', 'I_GROUP', 'LOC_QUOTIENT', 'PCT_RPT', 'OWN_CODE', 'OCC_CODE', 'H_PCT75', 'H_PCT90', 'A_PCT10','A_PCT25', 'A_PCT75', 'A_PCT90'], axis=1)
wage_estimates.head()
```

### Step 2: 

```python
# Rename the columns
wage_estimates = wage_estimates.rename(columns={
    'AREA_TITLE': 'State',
    'NAICS_TITLE': 'Industry',
    'OCC_TITLE': 'Occupation',
    'H_MEAN': 'Average hourly wage',
    'A_MEAN': 'Average annual wage',
    'H_MEDIAN': 'Median hourly wage',
    'A_MEDIAN': 'Median annual wage'
})

wage_estimates.head()
```

### Step 3: 

```python
#show only the rows related to California
california_data = wage_estimates[wage_estimates['State'] == 'California']
california_data.head()
```

### Step 4: 

```python
# show only the rows related to Data Scientists in California
data_scientists_ca = wage_estimates[(wage_estimates['State'] == 'California') & (wage_estimates['Occupation'] == 'Data Scientists')]
data_scientists_ca.head()
```
